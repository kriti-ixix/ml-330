## Style and Color
We've shown a few times how to control figure aesthetics in seaborn, but let's now go over it formally:

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
tips = sns.load_dataset('tips')
```

### Styles
You can set particular styles:

```python
sns.countplot(x= tips['sex'])
```

```python
sns.set_style('dark')
sns.countplot(x= tips['sex'])
```

```python
sns.set_style('ticks')
sns.countplot(x= tips['sex'], palette= 'deep')
```

### Spine Removal

```python
sns.countplot(x = 'sex', data= tips)
sns.despine()
```

```python
sns.countplot(x = tips['sex'])
sns.despine(left= True)
```

```python
plt.figure(figsize= (12,10))
sns.countplot(x= tips['day'])
```

```python
sns.lmplot(x='total_bill',y= 'tip', data= tips, aspect= 2)
```

## Scale and Context

The set_context() allows you to override default parameters

```python
sns.set_context('poster',font_scale= 1)
sns.countplot(x= tips['sex'], palette= 'coolwarm')
```
